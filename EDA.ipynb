```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import pandas as pd
import os
from datetime import datetime, timedelta, timezone
```

```python
# read csv files and set all variables as string
advertisement = pd.read_csv('advertisement.csv', dtype=str)
apartment = pd.read_csv('apartment.csv', dtype=str)
house = pd.read_csv('house.csv', dtype=str)
land = pd.read_csv('land.csv', dtype=str)
project = pd.read_csv('project.csv', dtype=str)
seller = pd.read_csv('seller.csv', dtype=str)
office = pd.read_csv('office.csv', dtype=str)
```

```python
columns_dict = {
    'advertisement': advertisement.columns,
    'apartment': apartment.columns,
    'house': house.columns,
    'land': land.columns,
    'project': project.columns,
    'seller': seller.columns,
    'office': office.columns
}
```

```python
print(columns_dict['advertisement'])
```

### Khảo sát apartment

```python
# print all the columns of apartment
print(columns_dict['apartment'])
```

```python
print(len(advertisement[advertisement['Diện tích'] != "None m²"]))
```

```python
total_record = len(apartment)
print(f'Total records: {total_record}')
# Chuyển giá bán về dạng số; giá trị lỗi sẽ thành <NA>
apartment['Giá bán'] = pd.to_numeric(
    apartment['Giá bán'].str.replace(r'[^0-9]', '', regex=True),
    errors='coerce'
).astype('Int64')

# Loại hình
print(apartment['Loại hình'].unique())

# Tình trạng nội thất
print(apartment['Tình trạng nội thất'].unique())

# Đặc điểm căn hộ
print(apartment['Đặc điểm căn hộ'].unique())

# Hình thức
print(apartment['Hình thức'].unique())

# Loại hình căn hộ
print(apartment['Loại hình căn hộ'].unique())

# Giấy tờ pháp lý
print(apartment['Giấy tờ pháp lý'].unique())

price_range = (apartment['Giá bán'].min(), apartment['Giá bán'].max())
min_price = apartment['Giá bán'].idxmin()
max_price = apartment['Giá bán'].idxmax()
print(f'Price range: {price_range}')
print(f'Lowest price: {apartment.loc[min_price]["Giá (chữ)"]}')
print(f'Highest price: {apartment.loc[max_price]["Giá (chữ)"]}')


# Ngày từ string sang datetime
apartment['Ngày đăng'] = pd.to_datetime(apartment['Ngày đăng'], errors='coerce')
datetime_range = (apartment['Ngày đăng'].min(), apartment['Ngày đăng'].max())
print(f'Datetime range: {datetime_range}')

# Tách phần số trong diện tích; giá trị không hợp lệ sẽ thành NaN
apartment['Diện tích'] = pd.to_numeric(
    apartment['Diện tích']
        .fillna('')
        .str.strip()
        .str.replace(',', '', regex=False)
        .str.extract(r'(\d+(?:\.\d+)?)', expand=False),
    errors='coerce'
)
area_range = (apartment['Diện tích'].min(), apartment['Diện tích'].max())
print(f'Area range: {area_range}')
```
